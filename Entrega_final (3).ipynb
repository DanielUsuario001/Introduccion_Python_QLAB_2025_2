## Sección 1: NumPy - Manipulación Avanzada de Arrays

**1.1.** Crea un array de NumPy llamado `vector_base` que contenga una secuencia de números desde 10 hasta 115 (inclusive), con un paso de 3 entre cada número.

```python
import numpy as np

# Crear el array
vector_base = np.arange(10, 116, 3)  # El 116 es para incluir el 115

# Mostrar el array
print(vector_base)
```

**1.2.** A partir del `vector_base` anterior, crea una matriz de 5x7 llamada `matriz_A`.

```python
# Paso anterior pero dejando solo los primeros 35 elementos
vector_base = np.arange(10, 116, 3)[:35]

# Creamos la matriz 5x7
matriz_A = vector_base.reshape(5, 7)

print(matriz_A)
```

**2.1.** Crea una segunda matriz de 5x7 llamada `matriz_B` que contenga números enteros aleatorios entre 1 y 10.

**2.2.** Realiza una multiplicación elemento por elemento (element-wise) entre `matriz_A` y `matriz_B` y almacena el resultado en `matriz_C`.

```python
matriz_B = np.random.randint(1, 11, size=(5, 7))

print(matriz_B)
```

```python
# Suponiendo que ya tenemos matriz_A y matriz_B
matriz_C = matriz_A * matriz_B  # Multiplicación elemento por elemento

print(matriz_C)
```

**3.1.** Utilizando slicing, selecciona una submatriz de 2x2 de la esquina inferior derecha de `matriz_C`.

**3.2.** Calcula el promedio de esta submatriz.

```python
submatriz = matriz_C[-2:, -2:]

print(submatriz)
```

```python
promedio = np.mean(submatriz)

print(promedio)
```

**7.** Crea una nueva columna condicional `Categoria_Desarrollo`. Clasifica a los países según las siguientes reglas:

**9.** Utilizando `groupby()`, calcula para cada `Continente`:
- La **media** de `Esperanza_de_vida`.

**14.1.** Calcula la matriz de correlación para las columnas numéricas del DataFrame.

**14.2.** **Interpretación:** Identifica el par de variables con la correlación positiva más fuerte y el par con la correlación negativa más fuerte (excluyendo la correlación de una variable consigo misma). ¿Tienen sentido estos resultados en el mundo real?

```python
# matriz de correlaciones (solo columnas numéricas)
num = df_desarrollo.select_dtypes(include='number')
corr = num.corr()  # Pearson por defecto
corr.round(3)    # <- se muestra la matriz
```

```python
# pares con mayor correlación
upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))

# Par con correlación positiva más fuerte
par_pos_max = upper.stack().idxmax()
val_pos_max = upper.stack().max()

# Par con correlación negativa más fuerte
par_neg_min = upper.stack().idxmin()
val_neg_min = upper.stack().min()

print(f"Mayor correlación positiva: {par_pos_max} = {val_pos_max:.3f}")
print(f"Mayor correlación negativa: {par_neg_min} = {val_neg_min:.3f}")
```

*Relaciones positivas fuertes esperadas: economías más ricas (PBI per capita alto) suelen exhibir mayor esperanza de vida y niveles educativos más altos, por lo que es razonable encontrarnos con una correlación alta entre las variables PBI_per_capita y Esperanza_de_vida*

*Relaciones negativas fuertes esperadas: mayor desigualdad (Gini alto) tiende a asociarse con peores indicadores sociales, como el PBI per capita. Por tanto, es razonable encontrarse con esta relación.*
